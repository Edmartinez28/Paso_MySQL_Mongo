```python
lista = []
for e in range(len(datos)):
    insert = 'db.encuestas.insertOne({'
    if datos.IdPersona[e] != 'none': 
        insert += 'IdPersona:"'+str(datos.IdPersona[e])+'"'
    if datos.Area[e] != 'none': 
        insert += ',Area:"'+str(datos.Area[e])+'"'
    if datos.Genero[e] != 'none': 
        insert += ',Genero:"'+str(datos.Genero[e])+'"'
    if datos.Edad[e] != 'none': 
        insert += ',Edad:'+str(datos.Edad[e])+''
    if datos.EstadoCivil[e] != 'none': 
        insert += ',EstadoCivil:"'+str(datos.EstadoCivil[e])+'"'
    if datos.Ingresos[e] != 'none': 
        insert += ',Ingresos:'+str(datos.Ingresos[e])+''
    if datos.Sectorizacion[e] != 'none': 
        insert += ',Sectorizacion:"'+str(datos.Sectorizacion[e])+'"'
    if datos.FactorExpansion[e] != 'none': 
        insert += ',FactorExpansion:'+str(datos.FactorExpansion[e])+''
    if datos.CondicionAct[e] != 'none': 
        insert += ',CondicionActividad:"'+str(datos.CondicionAct[e])+'"'
    if datos.Etnia[e] != 'none': 
        insert += ',Etnia:"'+str(datos.Etnia[e])+'"'
    if datos.NivelInst[e] != 'none': 
        insert += ',NivelInst:"'+str(datos.NivelInst[e])+'"'
    if datos.Provincia[e] != 'none': 
        insert += ',Provincia:"'+str(datos.Provincia[e])+'"'
    if datos.Zona[e] != 'none': 
        insert += ',Zona:'+str(datos.Zona[e])+''
    if datos.RamaAct[e] != 'none': 
        insert += ',RamaActividad:"'+str(datos.RamaAct[e])+'"'
    if datos.GrupoOcu[e] != 'none': 
        insert += ',GrupoOcupacion:"'+str(datos.GrupoOcu[e])+'"'
    
    insert += '})'
    lista.append(insert)    
lista
```
